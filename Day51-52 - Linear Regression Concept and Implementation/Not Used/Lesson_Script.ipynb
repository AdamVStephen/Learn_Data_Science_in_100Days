#### Predicting using Linear Regression

```python
# Removing cases where there was no price
housing_data_filtered = housing_data[housing_data.Price.notnull()]
housing_data_filtered.head()
```

```python
from pandas.api.types import is_string_dtype
#Tranform categorical data into numerical data for training purpose
for col, col_data in housing_data_filtered.items():
    if is_string_dtype(col_data):
        housing_data_filtered[col] = housing_data_filtered[col].astype('category').cat.as_ordered().cat.codes
```

```python
housing_data_filtered.head()
```

```python
corrmat = housing_data_filtered.corr()
corrmat.head(21)
```

```python
important_cols = ['Rooms', 'Type', 'Distance', 'Bedroom2', 'Bathroom', 'Car', 'YearBuilt', 'Price']
corrmat_v2 = housing_data_filtered[important_cols].corr()
corrmat_v2.head(8)
```

```python
#Feature Selection
important_cols.remove('Bedroom2')
cols = important_cols
housing_data_filtered = housing_data_filtered[cols]
housing_data_filtered.head()
```

```python
from sklearn.impute import SimpleImputer
#Replace null data cells with means value using Simple Imputer
y = housing_data_filtered.Price
cols.remove('Price')
X = SimpleImputer().fit_transform(housing_data_filtered[cols])
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
#Split the data into training and validation data sets
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(train_X, train_y)
```

```python
predict_y = model.predict(val_X)
```

```python
#Plot the actual Price value against predicted Sale Price 
g = sns.jointplot(x= val_y, y= predict_y, kind='reg', xlim=(0,8500000), ylim=(0,8500000),
                  joint_kws={'line_kws':{'color':'darkorange'}})
g.set_axis_labels(xlabel='Actual Price', ylabel='Predicted Price')
```

```python
predict_y_df = pd.DataFrame(predict_y)
compare = pd.concat([val_y.reset_index(drop=True), predict_y_df], axis=1)
compare.columns = ['Actual', 'Predicted']
compare['Actual'] = compare['Actual'].astype('int64')
compare['Predicted'] = compare['Predicted'].astype('int64')
compare.head()
```
