```python
# DISCLAIMER: 'df' refers to the data you passed in when calling 'dtale.show'

import numpy as np
import pandas as pd

duplicates = df.duplicated(['total'], keep='first')
df = df[~duplicates]
if isinstance(df, (pd.DatetimeIndex, pd.MultiIndex)):
	df = df.to_frame(index=False)

# remove any pre-existing indices for ease of use in the D-Tale code, but this is not required
df = df.reset_index().drop('index', axis=1, errors='ignore')
df.columns = [str(c) for c in df.columns]  # update columns to strings in case they are numbers

df = df.rename(columns={'total': 'total_new'})
s = df['total_new']
s = s.replace({
	10.12: 20})
df.loc[:, 'total_new'] = s
s = df['total_new']
s = s.replace({
	10.8: 30.8})
df.loc[:, 'total_new'] = s
s = df['total_new']
s = s.replace({
	12: 24})
df.loc[:, 'total_new'] = s
df.loc[:, 'Alcohol_Speed'] = df['alcohol'] * df['speeding']
corr_cols = [
	'total_new', 'speeding', 'alcohol', 'not_distracted', 'no_previous', 'ins_premium', 'ins_losses', 'Alcohol_Speed'
]
corr_data = np.corrcoef(df[corr_cols].values, rowvar=False)
corr_data = pd.DataFrame(corr_data, columns=[corr_cols], index=[corr_cols])
corr_data.index.name = str('column')
corr_data = corr_data.reset_index()
```
