Objective: Learn about selecting the data required from the dataset



```python
#import pandas library and provide alias name as pd
import pandas as pd
```

1. Reading the data

```python
trend_data = pd.read_csv('Data/SearchTrendData.csv')
trend_data.head(5)
```

2a. Selecting a column in the dataset

```python
trend_data.eLearning
#trend_data.eLearning
```

2b. Alternate way to select a column

```python
trend_data['Week']
```

2c. Select a specific column and row

```python
#trend_data['Week'][0]
trend_data['Week'][10]
```

2d. Selecting using loc method

```python
print(trend_data.loc[:,'Week'])
```

2e. Selecting multiple columns

```python
print(trend_data.loc[:,['Week', 'DataScience']])
```

2f. Selecting few columns and specific rows using index

```python
print(trend_data.loc[[0,1,2],['Week', 'DataScience']])
```

2g. Selecting specific columns and a range of rows using index

```python
print(trend_data.loc[250:260,['Week', 'DataScience']])
# Print including 250 and 260 line
```

```python
print(trend_data.iloc[250:260,])
# Using iloc now the last element is not printed
```

2h. select a range of rows and all columns

```python
print(trend_data.loc[:10,:])
```

2i. Using negaive index

```python
trend_data.loc[-1:,:]
```

3. Conditional Selection

```python
trend_data.Week == '13/09/2020'
# Identifies the row that has the value
```

3a. Selecting a row

```python
# Now passing that to the data frame will print that row
trend_data.loc[trend_data.Week == '13/09/2020']
```

3b. Using a condition

```python
trend_data.loc[(trend_data.DataScience >= 30)]
```

3c. Using multiple Conditions

```python
trend_data.loc[(trend_data.DataScience >= 30) & (trend_data.MachineLearning >= 30)]
```

3d. using OR operator

```python
trend_data.loc[(trend_data.DataScience >= 30) | (trend_data.MachineLearning >= 30)]
```

3e. Identifies the rows that are null

```python
trend_data.loc[trend_data.DataScience.notnull()]
```

```python
trend_data.loc[trend_data.DataScience.isnull()]
```

4. Assign value

```python
trend_data['EdgeCompute'] = 45
```

```python
trend_data.head()
```

5. Re-Indexing / Chaning the index

```python
print(trend_data)
```

```python
print(trend_data.set_index("Week"))
```
