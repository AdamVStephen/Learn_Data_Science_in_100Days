Binning

- Binning is a way to group data and this can be used for both categorical and numerical features
- Binning helps in preventing the over-fit in the model


```python
import numpy as np 
import pandas as pd 
```

```python
train = pd.read_csv("../Data/Titanic/train.csv")
test = pd.read_csv("../Data/Titanic/test.csv")
```

```python
train.head()
```

```python
# check the original distribution
title_test = train.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
pd.DataFrame(title_test.value_counts())
```

```python
data = [train, test]
titles = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Other": 5}

for dataset in data:
    # extract the title
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
    dataset['Title'].replace(['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don'],['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs','Other','Other','Other','Mr','Mr','Mr'],inplace=True)
    # convert titles into numbers
    dataset['Title'] = dataset['Title'].map(titles)
    # filling NaN with 0, to get safe
    dataset['Title'] = dataset['Title'].fillna(0)
 
```

```python
# check the distribution
pd.DataFrame(train.Title.value_counts())
```

```python
train.boxplot(column=['Age'])
```

```python
data_train_test = pd.concat([train.assign(ind="train"), test.assign(ind="test")], ignore_index=True)
age_nan_pos = data_train_test[data_train_test["Age"].isnull()].index
for i in age_nan_pos:
    mean_age = data_train_test["Age"][((data_train_test["Title"] == data_train_test.iloc[i]["Title"]))].mean()
    data_train_test["Age"].iloc[i] = mean_age
```

```python
test, train = data_train_test[data_train_test["ind"].eq("test")], data_train_test[data_train_test["ind"].eq("train")]
```

```python
data = [train, test]
for dataset in data:
    dataset['Age'] = dataset['Age'].astype(int)
    dataset.loc[ dataset['Age'] <= 15, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 15) & (dataset['Age'] <= 30), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 30) & (dataset['Age'] <= 45), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 45) & (dataset['Age'] <= 60), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 60, 'Age'] = 4
```

```python
# check the distribution
pd.DataFrame(train.Age.value_counts())
```

```python
#pd.qcut(train['Fare'], 13, labels=np.arange(13) + 1)
pd.qcut(train['Fare'], q=5)
```

```python
# Using Quartile cut - Generally used to divide the data into equal sized bins
data = [train, test]
for dataset in data:
    dataset['Fare'] = pd.qcut(dataset['Fare'], 5, labels=np.arange(5) + 1)
```

```python
# check the distribution
pd.DataFrame(train.Fare.value_counts())
```
