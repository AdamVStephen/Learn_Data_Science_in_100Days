#### Scaling
When we have multiple numerical variabes/features it is normal for them to be in different ranges like the Age and Salary will not be in same range but when we use distance based algorithm then it will be an issue. Because the feature with larger range would be given more importance hence to avoid this proble we need scaling


Why do we need Scaling?
- The distance based algorithms like KNN or K-Means requires the features to be scaled

1. Normalization - The values will be scaled to a fixed range between 0 and 1. Before normalization it is better to handle the outliers as otherwise the presence of outliers will increase

2. Standardization - This scales the values taking into account the standard deviation, the mean values will become 0 and the values will fall on either side of 0. Outliers in the data still could be a problem so they need to be handled before standrdization

```python
import numpy as np 
import pandas as pd
```

```python
train = pd.read_csv("../Data/Titanic/train.csv")
data = train[['Age', 'Fare']]
data.head()
```

```python
data['normalized_age'] = (data['Age'] - data['Age'].min()) / (data['Age'].max() - data['Age'].min())
```

```python
print("Minimum", data['normalized_age'].min())
print("Maximum", data['normalized_age'].max())
```

```python
data['standardized_age'] = (data['Age'] - data['Age'].mean()) / data['Age'].std()
```

```python
print("Minimum", data['standardized_age'].min())
print("Maximum", data['standardized_age'].max())
```
