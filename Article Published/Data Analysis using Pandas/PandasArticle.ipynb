```python
#import pandas library and provide alias name as pd
import numpy as np
import pandas as pd

trend_data = pd.read_csv('Data/SearchTrendData_ModifiedData_Original.csv')
print(trend_data.head())

trend_data_new = pd.read_csv('Data/SearchTrendData_ModifiedData_NewTopic.csv')
print(trend_data_new.head())
```

```python
# Filtering Specific data
trend_data.loc[(trend_data.Month == 10) & (trend_data.Year == 2019)]
```

```python
trend_data.columns = ['Week', 'Month', 'Year', 'eLearning', 'DataScience', 'MachineLearning', 'ArtificialIntelligence']
trend_data.head(10)
```

```python
trend_data.replace( to_replace=-1, method='ffill', inplace = True)
trend_data.head(10)
```

```python
trend_data.fillna(method='backfill', inplace = True)
trend_data.head(10)
```

```python
new_data = pd.merge(trend_data, trend_data_new, on='Week')
del new_data['Month_y']
del new_data['Year_y']
new_data.head()
```

```python
trend_data1 = trend_data
del trend_data1['Year']
```

```python
trend_data1.groupby(['Month']).mean()
```

```python
pd.crosstab(trend_data['Month'], trend_data['Year'])
#trend_data.head()
```

```python
del trend_data1['Month']
```
