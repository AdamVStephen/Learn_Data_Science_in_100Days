Topics Covered
- Rename - To rename a index or a column
- Replace - To replace a value with a new value

```python
#import pandas library and provide alias name as pd
import pandas as pd
import numpy as np

trend_data = pd.read_csv('Data/SearchTrendData.csv')
trend_data.head()
```

1a. Rename a Column of a dataframe

```python
td1 = trend_data.rename(columns={'MachineLearning': 'ML'})
td1.head()
```

1b. Rename index

```python
td2 = trend_data.rename(index={0: 'a', 1: 'b'})
td2.head()
```

1c. Giving a name to axis

```python
td3 = trend_data.rename_axis("Dates", axis='rows').rename_axis("Topics", axis='columns')
td3.head()
```

1d. Renaming by passing the new values

```python
td4 = trend_data
print(trend_data)
td4.columns = ['Wk', 'eL', 'DS', 'ML', 'AI', 'DL']
print(td4)
```

1e. Replace a default string in the header

```python
td5 = trend_data
td5.columns = ['$Week', '$eLearn', '$DS', '$ML', 'AI', 'DL']
print(td5)
td5.columns = td5.columns.str.replace('$','')
print(td5)
```

2a. Replace using Loc

```python
t_data = trend_data[-10:]
t_data = t_data.reset_index(drop=True) # For resetting index
print(t_data)
```

```python
# Using Loc
print(t_data.loc[0, 'DataScience'])
```

```python
t_data.loc[0, 'DataScience'] = 40
t_data.head()
```

2b. Replace using At

```python
# Using At
t_data.at[1, 'DataScience'] = 40
t_data.head()
```

2c. Using replace

```python
t_data.replace(40, 42) 
```

```python
t_data.head() # Original dataframe has not changed
```

2d. Replace with inplace

```python
t_data.replace(40, 42, inplace = True) # Original dataframe has changed
t_data.head()
```

2e. Replace a list of items

```python
t_data.replace(to_replace=[42, 21, 15],
           value= [142, 121, 115])
```

```python
trend_data = pd.read_csv('Data/SearchTrendData_WithMissing.csv')
t_data_m = trend_data[:20]
t_data_m = t_data_m.reset_index(drop=True) # For resetting index
print(t_data_m)
```

2f. Replace with forwardfill and backfill

```python
t_data_m.replace(to_replace=35.0, value=None, method='ffill')
```

```python
t_data_m.replace(to_replace=35.0, value=None, method='bfill')
```
