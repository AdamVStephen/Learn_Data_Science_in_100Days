Topics to be Covered
- Central limit Theorem: According to the CLT, as we take more samples from a distribution, the sample averages will tend towards a normal distribution regardless of the population distribution.
- Margin of Error

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
```

```python
data = pd.read_csv('../Data/clt_data.csv')
print(data.head())
```

```python
plt.hist(data.Weight, density=True, bins=30) 
```

```python
def random_subsets(data, tot_subsets, sub_size):
    subset_means = []
    for i in range(tot_subsets):
        subset = data.sample(n=sub_size)
        subset_mean = np.array(subset).mean()
        subset_means.append(subset_mean)
    return subset_means
```

```python
subsets_from_data = random_subsets(data, 100, 50) 
# Keep increasing the total number of subsets
sns.distplot(subsets_from_data, hist=False)
```

```python
data2 = pd.read_csv('../Data/clt_data2.csv')
plt.hist(data2.Weight, density=True, bins=30) 
```

```python
subsets_from_data2 = random_subsets(data2, 5000, 30) 
sns.distplot(subsets_from_data2, hist=False)
```
